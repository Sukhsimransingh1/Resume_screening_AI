```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

```python
skills_list = [
    "python", "machine learning", "sql", "excel", "react",
    "java", "spring boot", "fastapi", "data analysis",
    "pandas", "numpy", "power bi"
]
```

```python
def extract_skills(text):
    text = text.lower()
    found_skills = []
    
    for skill in skills_list:
        if skill in text:
            found_skills.append(skill)
    
    return list(set(found_skills))
```

```python
resume = "Python developer with machine learning, pandas and numpy experience"
jd = "Looking for Python, machine learning and SQL skills"
```

```python
resume_skills = extract_skills(resume)
jd_skills = extract_skills(jd)
```

```python
missing_skills = list(set(jd_skills) - set(resume_skills))

print("Resume Skills:", resume_skills)
print("JD Skills:", jd_skills)
print("Missing Skills:", missing_skills)
```
